```python
import pandas as pd #this imports the pandas as pd

df = pd.read_csv('/content/GlobalLandTemperaturesByCity.csv') #this reads the csv file

df['dt'] = pd.to_datetime(df['dt']) #this converts the dt column to dataframe
start = '-01-01' #this gives the range for filtering out the data
end = '-12-31' #this gives the range for filtering out the data
list_dates = [] #this initializes the list to store the dates
us_temp = [] #this is the average temperature for united states
in_temp = [] #this is the average temperature for india
for date in range(1743, 2014, 3): #this creates a loop of years from 1743 to 2014 with a interval of 3
    date = str(date) #this converts the date to a string
    start_date = date + start #this basically gives the start date
    end_date = date + end #this basically gives the end date
    in_data = df[(df['Country'] == 'India') & (df['dt'] >= start_date) & (df['dt'] <= end_date)] #this filters the data for India within the specific interval
    in_data = in_data['AverageTemperature'].dropna() #this filters out the Nan values
    in_mean = in_data.mean() if not in_data.empty else 0 #this calculates the mean or sets the data to zero if there's no data
    in_temp.append(in_mean) #this appends the mean temperature to the list for India
    us_data = df[(df['Country'] == 'United States') & (df['dt'] >= start_date) & (df['dt'] <= end_date)] #this filters data for United States within the specific interval
    us_data = us_data['AverageTemperature'].dropna() #this filters out the Nan values
    us_mean = us_data.mean() if not us_data.empty else 0 #this calculates the mean or sets the data to zero if there's no data
    us_temp.append(us_mean) #this appends the mean temperature to the list for United States
    list_dates.append(date) #this adds the year to the list of years
print("United States:", us_temp) #this prints the average for United States
print("India:", in_temp) #this prints the average for India
```

```python
# GRAPH NUMBER ONE

import pandas as pd #pandas is invaluable to anyone using csv files as without pandas it is impossible to use a csv file, we also import it as pd as it makes writing any code with pandas just a few characters shorter.
from collections import Counter #counter is used to count up the amount of instances of a value appear in a list or csv file.
import re #this lets us search for special characters and remove them without trouble.
import matplotlib.pyplot as plt #this is used for ploting line graphs, bar graphs, and ex.

d5 = pd.read_csv('/content/GlobalLandTemperaturesByCity.csv') #applys Global land temperatures by city csv file to the variable d5, making the implementation and editing of the csv file easier and leaves the orginal csv file unedited.

def lower_and_split(value): #this entire def is for the lowering and spliting of the countrys, as some of the countrys make have an apostrophe while others do not but the counter will consider this two words as different words
    if pd.isna(value): #by removing the the ability for this to happen we will get a more accurate number of occurrences that this word appears in the csv file
        return [] #
    value = value.lower() #this converts the text to lowercase
    return [lang.strip() for lang in re.split(r'[,\/]+', value)]

all_countries = d5['Country'].dropna().tolist() #applies all of the country values in the csv file to all_countries named that way because it contains all of the countries
all_countries = [item for value in all_countries for item in lower_and_split(value)] #puts all_countries through lower_and_split which cleans up the csv file.
country_counter = Counter(all_countries) #this uses the counter code to count up all of the occurrences of each word and puts each one into a clear total
most_common_country, country_count = country_counter.most_common(1)[0] #this identifies the most common country and the occurrences of it

all_cities = d5['City'].dropna().tolist() #this basically extracts the city column, leaves the missing values and then converts it to a list
all_cities = [item for value in all_cities for item in lower_and_split(value)] #this basically splits the city names using the same function
city_counter = Counter(all_cities) #this counts the occurrences of each city name using using Counter
most_common_city, city_count = city_counter.most_common(1)[0] #this identifies the most common city and occurrence count of it

labels = ['Most Common Country', 'Most Common City'] #this is the labels for the bar graph
counts = [country_count, city_count] #this correlates for each label

plt.figure(figsize=(6, 4)) #this basically sets the figure size so that we can better read the graph
plt.bar(labels, counts, color=['blue', 'red']) #this creates a bar graph with different colors
plt.title("Most Common Country & Most Common City") #this adds a tittle to the graph
plt.xlabel("Category") #this gives the label for the x-axis
plt.ylabel("Occurrences") #this gives the label for the y-axis
plt.show() #this displays the graph

print(f"The most common country is '{most_common_country}' with {country_count} occurrences.") #this prints the number of occurrences for the most common country
print(f"The most common city is '{most_common_city}' with {city_count} occurrences.") #this prints the number of occurrences for the most common city
```

```python
# GRAPH NUMBER TWO

import pandas as pd #this imports the pandas as pd
import matplotlib.pyplot as plt #this is used for ploting line graphs, bar graphs, and ex

d5 = pd.read_csv('/content/GlobalLandTemperaturesByCity.csv') #this reads the csv file

d5['dt'] = pd.to_datetime(d5['dt']) #this basically converts the dt column to datetime for the formatting purpose

start_date_1 = '1743-10-01' #this is the start date of the first interval
end_date_1 = '1956-03-01' #this is the end date of the second interval

start_date_2 = '1956-04-01' #this is start date of the second interval
end_date_2 = '2023-11-11' #this is the end date of the second interval

filtered_df_1 = d5[(d5['dt'] >= start_date_1) & (d5['dt'] <= end_date_1)].dropna(subset = ['AverageTemperature']) #this filters the dataframe so that it would include rows within the first range and it would exclude the Nan values
filtered_df_2 = d5[(d5['dt'] >= start_date_2) & (d5['dt'] <= end_date_2)].dropna(subset = ['AverageTemperature']) #this filters the dataframe so that it would include rows within the second range and it would exclude the Nan values

mean_1 = filtered_df_1['AverageTemperature'].mean() #this calculates the mean temperature for the first interval
mean_2 = filtered_df_2['AverageTemperature'].mean() #this calculates the mean temperature for the second interval

temp_diff = abs(mean_1 - mean_2) #this basically just calculates the absoulte difference between the two mean temperatures

print(f"Mean Temperature between {start_date_1} and {end_date_1}: {mean_1}") #this prints the mean temperature for the first interval
print(f"Mean Temperature between {start_date_2} and {end_date_2}: {mean_2}") #this prints the mean temperature for the second interval
print(f"Temperature Difference: {temp_diff}") #this prints the temperature difference between those two intervals

intervals = [f'{start_date_1} to {end_date_1}', f'{start_date_2} to {end_date_2}', 'Difference'] #this basically gets the data for the graph
mean_temp = [mean_1, mean_2, temp_diff] #this gives the mean temperature of the intervals

plt.figure(figsize = (10, 6)) #this basically sets the figure size so that we can better read the graph
plt.bar(intervals, mean_temp, color=['blue', 'green', 'red']) #this creates a bar graph with different colors
plt.xlabel('Date Intervals') #this gives the label for the x-axis
plt.ylabel('Mean Temperature') #this gives the label for the y-axis
plt.title('Comparison of Mean Temperatures Between Intervals and Their Difference') #this adds a tittle to the graph
plt.show() #this displays the graph
```

```python
# UNIQUE GRAPGH NUMBER 11

#https://www.statology.org/pandas-read_csv-usecols/
#https://stackoverflow.com/questions/49673345/python-pandas-get-rows-of-a-dataframe-where-a-column-is-not-null
#https://stackoverflow.com/questions/46938572/pandas-groupby-mean-into-a-dataframe
#https://www.educative.io/answers/how-to-create-a-choropleth-map-with-plotly-express-in-python
#https://stackoverflow.com/questions/56709920/list-of-locationmode-to-use-with-plotly
#https://stackoverflow.com/questions/59953431/how-to-change-plotly-figure-size

import pandas as pd #this imports the pandas as pd
import plotly.express as px #this imports plotly for the plotting

df = pd.read_csv('GlobalLandTemperaturesByCity.csv', usecols = ['Country', 'AverageTemperature']) #this basically just reads the perticular columns which are "Country" and "AverageTemperature" from this csv file

df = df[df['AverageTemperature'].notnull()] #this basically just filters out the data to include rows where this perticular column - "AverageTemperature" is empty

df = df.groupby('Country', as_index = False).mean() #this basically just groub the data by the column - "Country" so this would calculate the mean of each group in this case different countries

fig = px.choropleth(df, #this creates a choropleth map so that we can visualize the average temperature of each country in this globe
                    locations = "Country", #this selects this column of "Country" so that it would select all of the countries in this csv file
                    locationmode = "country names", #this basically just specifies the format of the location data
                    color = "AverageTemperature", #this column is basically used to determine the intensity of the color
                    hover_name = "Country", #this column is basically to display the hovering label
                    title = "World Temperature Comparison", #this basically gives the title of the choropleth map
                    labels = {"AverageTemperature": "Avg Temperature (°C)"}, #this gives the label for the color bar on the right side
                    color_continuous_scale = "Viridis") #this basically just specifies the color for the scale

fig.update_layout( #this is basically just adjusting the layout of the map
    autosize = False, #this is disabling the automatic resizing
    width = 1100, #this is setting the width of the figure
    height = 1100, #this is setting the height of the figure
)

fig.show() #this basically just displays the figure
```
